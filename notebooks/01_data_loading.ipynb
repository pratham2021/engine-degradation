```python
# Fan section (front of engine)
# sensor1 -- Total temperature at fan inlet -- air temperature entering the engine
# sensor2 -- Total temperature at fan outlet -- temperature after passing through the fan
# sensor3 -- Total temperature at Low Pressure Compressor (LPC) outlet
# sensor4 -- Total temperature at High Pressure Compressor (HPC) outlet -- one of the most important degradation indicators
```

```python
# Compressor section
# sensor5 -- Total temperature at bypass duct
# sensor6 -- Total temperature at High Pressure Tubine (HPT) inlet
# sensor7 -- Total pressure at HPC outlet -- air pressure after maximum compression
# sensor8 -- Physical fan speed (RPM) -- how fast the front fan is spinning
# sensor9 -- Physical core speed (RPM) -- how fast the engine core is spinning
```

```python
# Turbine and pressure section
# sensor10 -- Pressure ratio across the engine
# sensor11 -- Static pressure at HPC outlet
# sensor12 -- Fuel flow ratio -- how much fuel is being burned
# sensor13 -- Corrected fan speed -- fan speed adjusted for atmospheric conditions
# sensor14 -- Corrected core speed -- core speed adjusted for atmospheric conditions
```

```python
# Efficiency and bypass section
# sensor15 -- Bypass ratio -- ratio of air going around the core vs through it
# sensor16 -- Burner fuel-air ratio
# sensor17 -- Bleed enthalpy -- energy in air bled off the compressor
# sensor18 -- Demanded fan speed
# sensor19 -- Demanded corrected fan speed
```

```python
# Coolant section
# sensor20 -- HPT coolant bleed -- cooling air going to the high pressure turbine
# sensor21 -- LPT coolant bleed -- cooling air going to the low pressure turbine
```
